The regressions below suggest that, on average, there is a positive correlation between average income permonth and charged price per night.  But the benefit might begin to taper off after about 400 dollars per night.

```python
plt.figure(figsize = (23,22))
plt.subplot(3,1,1)
#sns.scatterplot(data = airbnb, x='price', y ='middle_estimate/month', alpha = .4, color = 'blue')
sns.regplot(data = airbnb, x='price', y ='middle_estimate/month')
plt.subplot(3,1,2)
sns.regplot(data = airbnb[(airbnb['price']<=300) & (airbnb['middle_estimate/month'] <=5000)], x='price', y ='middle_estimate/month')
plt.subplot(3,1,3)
sns.regplot(data = airbnb[(airbnb['price']>=400) & (airbnb['middle_estimate/month'] <=10000)], x='price', y ='middle_estimate/month')
```

```python
for col in airbnb.columns:
    if len(airbnb[col].unique()) == 1:
        airbnb.drop(col,inplace=True,axis=1)
```

```python
print(airbnb.shape)
```

```python
airbnb.info()
```

```python
corr=airbnb.drop(columns = ['price','middle_estimate/month','lowest_estimate/month']).corr().abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]
print(to_drop)
```

```python
airbnb = airbnb.drop(columns = to_drop)
airbnb.shape
```

```python
corr=airbnb.drop(columns = ['price']).corr().abs()
corr = corr.unstack().sort_values().reset_index()
corr.columns
corr = corr[corr['level_0']!=corr['level_1']]
```

```python
corr.tail()
```

```python
price_corr = airbnb.corr()
price_corr = price_corr['price'].dropna().sort_values().reset_index()
price_corr = price_corr.iloc[price_corr['price'].abs().argsort()]
print(price_corr.tail(20))
```

```python
midestimate_corr = airbnb.corr()
midestimate_corr = midestimate_corr['middle_estimate/month'].dropna().sort_values().reset_index()
midestimate_corr = midestimate_corr.iloc[midestimate_corr['middle_estimate/month'].abs().argsort()]
print(midestimate_corr.tail(20))
```

```python
lowestimate_corr = airbnb.corr()
lowestimate_corr = lowestimate_corr['lowest_estimate/month'].dropna().sort_values().reset_index()
lowestimate_corr = lowestimate_corr.iloc[lowestimate_corr['lowest_estimate/month'].abs().argsort()]
print(lowestimate_corr.tail(20))
```
